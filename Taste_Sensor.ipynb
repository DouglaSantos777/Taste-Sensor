```python
data_set2 = movies.drop(columns=['overview', 'genre'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf=TfidfVectorizer(max_features=10000, stop_words='english')
```

```python
tf
```

```python
vector=tf.fit_transform(data_set2['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
def recommand(movie_title):
    if movie_title not in data_set2['title'].values:
        print(f"O filme '{movie_title}' não foi encontrado no dataset.")
        return

    index = data_set2[data_set2['title']==movie_title].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])

    print(f"\nRecomendações baseadas em '{movie_title}':")
    for i in distance[1:6]:
        title = data_set2.iloc[i[0]].title
        score = i[1]
        print(f"- {title}")
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
genres_list = movies['genre'].dropna().str.split(',', expand=True).stack()
genres_list = genres_list.str.strip()
top_genres = genres_list.value_counts().nlargest(15)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_genres.index, y=top_genres.values, palette="viridis")
plt.title('Distribuição dos 15 Gêneros Mais Frequentes')
plt.xlabel('Gênero')
plt.ylabel('Contagem de Filmes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
sample_size = 20
similarity_sample = similarity[:sample_size, :sample_size]

sample_titles = data_set2['title'].head(sample_size)

plt.figure(figsize=(10, 8))
sns.heatmap(similarity_sample,
            annot=False,
            cmap='plasma',
            fmt=".2f",
            xticklabels=sample_titles,
            yticklabels=sample_titles)
plt.title(f'Mapa de Calor da Similaridade de Cosseno (Amostra de {sample_size} Filmes)')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
filme_origem = "Wrong Turn"
index = data_set2[data_set2['title'] == filme_origem].index[0]
distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])

top_scores = distances[1:6]

recommended_titles = [data_set2.iloc[i[0]].title for i in top_scores]
similarity_scores = [i[1] for i in top_scores]

plt.figure(figsize=(10, 6))
sns.barplot(x=similarity_scores, y=recommended_titles, palette="magma")
plt.title(f'Força da Similaridade para o filme: "{filme_origem}"')
plt.xlabel('Pontuação de Similaridade de Cosseno')
plt.ylabel('Filmes Recomendados')
plt.xlim(0, 1)
plt.tight_layout()
plt.show()
```
